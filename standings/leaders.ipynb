```python
import time
from nba_api.stats.endpoints import commonallplayers, playergamelog
```

```python
# define season of interest
season = '2024-25'
# get list of all active players
allPlayers = commonallplayers.CommonAllPlayers(0, season=season, timeout=60).get_dict()
activePlayers = []
for player in allPlayers['resultSets'][0]['rowSet']:
	if player[3] == 1:
		activePlayers.append(player)
```

```python
playerAverages = {}
for player in activePlayers:
	time.sleep(3.5)
	gameLog = playergamelog.PlayerGameLog(player[0], timeout=60).get_dict()
	playerStats = [0] * 12
	for game in gameLog['resultSets'][0]['rowSet']:
		playerStats[0] += game[24] # points
		playerStats[1] += game[19] # assists
		playerStats[2] += game[18] # rebounds
		playerStats[3] += game[20] # steals
		playerStats[4] += game[21] # blocks
		playerStats[5] += game[22] # turnovers
		playerStats[6] += game[7] # fgm
		playerStats[7] += game[8] # fga
		playerStats[8] += game[10] # 3pm
		playerStats[9] += game[11] # 3pa
		playerStats[10] += game[13] # ftm
		playerStats[11] += game[14] # fta

	# fga = 0
	if playerStats[7] == 0:
		playerStats[7] = 1
	if playerStats[9] == 0:
		playerStats[9] = 1
	if playerStats[11] == 0:
		playerStats[11] = 1
	gamesPlayed = len(gameLog['resultSets'][0]['rowSet'])
	if gamesPlayed == 0:
		gamesPlayed = 1
	playerAverages[player[2]] = {
		'points':round(playerStats[0] / gamesPlayed, 2),
		'assists':round(playerStats[1] / gamesPlayed, 2),
		'rebounds':round(playerStats[2] / gamesPlayed, 2),
		'steals':round(playerStats[3] / gamesPlayed, 2),
		'blocks':round(playerStats[4] / gamesPlayed, 2),
		'turnovers':round(playerStats[5] / gamesPlayed, 2),
		'fg%':round((playerStats[6] / playerStats[7]) * 100, 2),
		'3fg%':round((playerStats[8] / playerStats[9]) * 100, 2),
		'ft%':round((playerStats[10] / playerStats[11]) * 100, 2),
		'games':gamesPlayed
	}
	print(player[2])
	print(playerAverages[player[2]])
```

```python
stats = sorted(playerAverages.items(), key=lambda x: x[1]['points'], reverse=True)
print('POINTS LEADERS')
for i in range(5):
	print(stats[i][0], '=>', stats[i][1]['points'])
stats = sorted(playerAverages.items(), key=lambda x: x[1]['assists'], reverse=True)
print('\nASSIST LEADERS')
for i in range(5):
	print(stats[i][0], '=>', stats[i][1]['assists'])
stats = sorted(playerAverages.items(), key=lambda x: x[1]['rebounds'], reverse=True)
print('\nREBOUND LEADERS')
for i in range(5):
	print(stats[i][0], '=>', stats[i][1]['rebounds'])
stats = sorted(playerAverages.items(), key=lambda x: x[1]['steals'], reverse=True)
print('\nSTEAL LEADERS')
for i in range(5):
	print(stats[i][0], '=>', stats[i][1]['steals'])
stats = sorted(playerAverages.items(), key=lambda x: x[1]['blocks'], reverse=True)
print('\nBLOCK LEADERS')
for i in range(5):
	print(stats[i][0], '=>', stats[i][1]['blocks'])
stats = sorted(playerAverages.items(), key=lambda x: x[1]['turnovers'], reverse=True)
print('\nTURNOVER LEADERS')
for i in range(5):
	print(stats[i][0], '=>', stats[i][1]['turnovers'])
```
